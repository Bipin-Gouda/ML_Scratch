```python
def prob(df1,f1,f2,f3,f4,output):
  totlen=len(df1)
  df1=df1[df1['Play Tennis']==output]
  prob_f1=len(df1[df1['Outlook']==f1])/len(df1)
  prob_f2=len(df1[df1['Temp']==f2])/len(df1)
  prob_f3=len(df1[df1['Humidity']==f3])/len(df1)
  prob_f4=len(df1[df1['Wind']==f4])/len(df1)
  prob=(len(df1)/totlen)*prob_f1*prob_f2*prob_f3*prob_f4
  return prob
```

```python
PYes=prob(df1,'Sunny','Cool','High','Strong','Yes')
PNo=prob(df1,'Sunny','Cool','High','Strong','No')
if PYes>PNo:
  print('Person plays tennis')
else:
  print('Person doesnt play tennis')
```
