# DATA Selection

```python
# DATA 1

x=list(data['Dataset 1'][1:6])
y=list(data['Unnamed: 1'][1:6])


# DATA 2
#x=list(data['Dataset 1'][11:27])
#y=list(data['Unnamed: 1'][11:27])
```

# Calculating Coefficients B1, B0 of (y=B0+B1x)

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
print(mean_x,mean_y)
def Coefficients(x,y):
  numerator=0
  denominator=0
  for i in range(len(x)):
    numerator=numerator+(x[i]-mean_x)*(y[i]-mean_y)
    denominator=denominator+(x[i]-mean_x)**2
    B1=numerator/denominator
    B0=mean_y-B1*mean_x
  return B1, B0

B1,B0=Coefficients(x,y)

print(B1,B0)
```

# Calculating Y_predicted

```python
def Predicted_Y(x):
  Y_predicted=[]
  for  i in range(len(x)):
    Y_predicted.append(B0+B1*x[i])
  return Y_predicted


Y_predicted=Predicted_Y(x)
print("Y_predicted is :",Y_predicted)
```

# Comparing Both the graphs in the same plot

```python
plt.scatter(x,y)
plt.plot(x,Y_predicted)
```

# Calculating RMSE

```python
#RMSE
def Rmse(y,Y_predicted):
  sum=0
  for i in range(len(y)):
    sum=sum+(Y_predicted[i]-y[i])**2
  MSE=sum/len(y)
  RMSE=MSE**0.5
  return RMSE

RMSE=Rmse(y,Y_predicted)
print(RMSE)
```
