```python
S=[[1,0,1],[3,1,1],[3,-1,1]]
coeff_matrix=[]
for i in range(len(S[0])):
  temp=[]
  for j in range(len(S)):
    temp.append(np.matmul(S[j],S[i]))
  coeff_matrix.append(temp)
print(coeff_matrix)
```
