# Stochastic Gradient Descent

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('/content/drive/MyDrive/Copy of Lab 8 ML Dataset - Stochastic Gradient Descent.xlsx')
```

```python
df2=df.iloc[25:35]
df2.columns=df.iloc[1]
df2.head()
```

```python
df2.shape
```

```python
df.columns=df.iloc[1]
df=df.iloc[2:22]
df.head()
```

we have 4 features here so we need four weights And a W0 (bias)

```python
W0=0.1
W1=0.2
W2=0.3
W3=0.4
W4=0.5
net=0
c=0.1
#dellk=0
for _ in range(4):
  for index, row in df.iterrows():
    net= W0*1+W1*row['X1'] + W2*row['X2'] + W3*row['X3'] + W4*row['X4']
    ypred=1/(1+np.exp(-net))       # fnet=ypred
    dellk=ypred*(1-ypred)*((row['Y']-ypred))    #f`(net)*(di-Oi)

    W1=W1+c*dellk*row['X1']              # W1+dellw1,     dellw=rcx1
    W2=W1+c*dellk*row['X2']
    W3=W3+c*dellk*row['X3']                  #r=(yi-ypred)
    W4=W4+c*dellk*row['X4']
    W0=W0+c*dellk*1                                 #x0=1  always

print(W0,W1,W2,W3,W4)
```

```python
# Now using these weights we will try to predict the test dataset
```

```python
probabilities=[]
y_pred=[]
for index, row in df2.iterrows():
  value=W0*1+W1*row['X1'] + W2*row['X2'] + W3*row['X3'] + W4*row['X4']
  probability = 1 / (1 + np.exp(-value))
  probabilities.append(probability)
  if probability>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
print(probabilities)
print(y_pred)
```
